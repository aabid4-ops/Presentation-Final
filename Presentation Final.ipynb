```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# ---------------------------------------------------------
# Embedded dataset (fully inside the code)
# ---------------------------------------------------------
DATA = [
    {"customer_id": 101, "age": 25, "country": "USA", "purchase_amount": 120.5, "items": 3, "days_since_signup": 10},
    {"customer_id": 102, "age": 32, "country": "Canada", "purchase_amount": 89.9, "items": 1, "days_since_signup": 50},
    {"customer_id": 103, "age": 41, "country": "UK", "purchase_amount": 150.0, "items": 4, "days_since_signup": 120},
    {"customer_id": 104, "age": 29, "country": "Germany", "purchase_amount": 60.0, "items": 2, "days_since_signup": 5},
    {"customer_id": 105, "age": 37, "country": "India", "purchase_amount": 200.0, "items": 5, "days_since_signup": 300},
    {"customer_id": 106, "age": 22, "country": "Brazil", "purchase_amount": 45.5, "items": 1, "days_since_signup": 40},
    {"customer_id": 107, "age": 55, "country": "USA", "purchase_amount": 300.0, "items": 6, "days_since_signup": 200},
    {"customer_id": 108, "age": 48, "country": "Canada", "purchase_amount": 110.0, "items": 2, "days_since_signup": 180},
    {"customer_id": 109, "age": 33, "country": "UK", "purchase_amount": 75.0, "items": 1, "days_since_signup": 90},
    {"customer_id": 110, "age": 27, "country": "Germany", "purchase_amount": 130.0, "items": 3, "days_since_signup": 60},
    {"customer_id": 111, "age": 44, "country": "India", "purchase_amount": 210.0, "items": 4, "days_since_signup": 250},
    {"customer_id": 112, "age": 39, "country": "Brazil", "purchase_amount": 55.0, "items": 1, "days_since_signup": 15},
    {"customer_id": 113, "age": 23, "country": "USA", "purchase_amount": 95.0, "items": 2, "days_since_signup": 30},
    {"customer_id": 114, "age": 31, "country": "Canada", "purchase_amount": 140.0, "items": 3, "days_since_signup": 100},
    {"customer_id": 115, "age": 46, "country": "UK", "purchase_amount": 160.0, "items": 4, "days_since_signup": 220},
    {"customer_id": 116, "age": 52, "country": "Germany", "purchase_amount": 190.0, "items": 5, "days_since_signup": 330},
    {"customer_id": 117, "age": 28, "country": "India", "purchase_amount": 80.0, "items": 1, "days_since_signup": 12},
    {"customer_id": 118, "age": 36, "country": "Brazil", "purchase_amount": 65.0, "items": 2, "days_since_signup": 45},
    {"customer_id": 119, "age": 49, "country": "USA", "purchase_amount": 250.0, "items": 5, "days_since_signup": 275},
    {"customer_id": 120, "age": 30, "country": "Canada", "purchase_amount": 100.0, "items": 2, "days_since_signup": 85},
]

# Convert to DataFrame
df = pd.DataFrame(DATA)

# ---------------------------------------------------------
# Visualization functions
# ---------------------------------------------------------
def plot_purchase_distribution(df):
    plt.figure(figsize=(8, 5))
    sns.histplot(df["purchase_amount"], bins=10, kde=True)
    plt.title("Purchase Amount Distribution")
    plt.xlabel("Purchase Amount")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()

def plot_age_vs_purchase(df):
    plt.figure(figsize=(8, 5))
    sns.scatterplot(data=df, x="age", y="purchase_amount", hue="country", s=100)
    plt.title("Age vs Purchase Amount")
    plt.xlabel("Age")
    plt.ylabel("Purchase Amount")
    plt.tight_layout()
    plt.show()

def plot_avg_purchase_by_country(df):
    grouped = df.groupby("country")["purchase_amount"].mean().reset_index()
    plt.figure(figsize=(8, 5))
    sns.barplot(data=grouped, x="country", y="purchase_amount")
    plt.title("Average Purchase Amount by Country")
    plt.xlabel("Country")
    plt.ylabel("Average Purchase Amount")
    plt.tight_layout()
    plt.show()

def plot_purchase_over_time(df):
    grouped = df.groupby("days_since_signup")["purchase_amount"].sum().reset_index()
    plt.figure(figsize=(10, 5))
    sns.lineplot(data=grouped, x="days_since_signup", y="purchase_amount")
    plt.title("Total Purchase Amount Over Time")
    plt.xlabel("Days Since Signup")
    plt.ylabel("Total Purchase Amount")
    plt.tight_layout()
    plt.show()

# ---------------------------------------------------------
# Run all graphs
# ---------------------------------------------------------
plot_purchase_distribution(df)
plot_age_vs_purchase(df)
plot_avg_purchase_by_country(df)
plot_purchase_over_time(df)
```
